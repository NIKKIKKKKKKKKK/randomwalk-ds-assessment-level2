```python
import pandas as pd
import numpy as np
import os

file_path = "dataset.csv"

if os.path.exists(file_path):
    try:
        df = pd.read_csv(file_path)

        print("Dataset Info:")
        df.info()
        print("\nMissing Values:\n", df.isnull().sum())

        duplicates = df.duplicated().sum()
        print("\nDuplicate Rows:", duplicates)

        if duplicates > 0:
            df = df.drop_duplicates()
            print("Duplicates removed.")

        df_cleaned = df.dropna()

        if 'Column1' in df.columns and df['Column1'].dtype in [np.float64, np.int64]:
            df['Column1'] = df['Column1'].fillna(df['Column1'].mean())
        else:
            print("'Column1' is missing or not numeric, skipping fillna for this column.")

        if 'Column2' in df.columns:
            df['Column2'] = df['Column2'].fillna('Unknown')
        else:
            print("'Column2' is missing, skipping fillna for this column.")

        print("\nColumn Data Types:\n", df.dtypes)

        df_cleaned.to_csv("cleaned_dataset.csv", index=False)
        print("Data Cleaning Completed. Cleaned dataset saved.")
    except Exception as e:
        print(f"An error occurred: {e}")
else:
    print(f"Error: File not found at '{file_path}'")
```

```python
import pandas as pd
from scipy.stats import skew, kurtosis

# Load the dataset
file_path = "dataset.csv"
df = pd.read_csv(file_path)

# Function to calculate skewness and kurtosis
def calculate_skew_kurtosis(group):
    return pd.Series({
        'bill_length_skew': skew(group['bill_length_mm'], nan_policy='omit'),
        'bill_length_kurtosis': kurtosis(group['bill_length_mm'], nan_policy='omit'),
        'bill_depth_skew': skew(group['bill_depth_mm'], nan_policy='omit'),
        'bill_depth_kurtosis': kurtosis(group['bill_depth_mm'], nan_policy='omit')
    })

# Group by species and apply the function, excluding the grouping column
skew_kurtosis_results = df.groupby('species', group_keys=False).apply(calculate_skew_kurtosis)

# Output results
print("Skewness and Kurtosis of bill_length_mm and bill_depth_mm by species:")
print(skew_kurtosis_results)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os  


file_path = "dataset.csv"

if os.path.exists(file_path):  
    df = pd.read_csv(file_path)

    
    numeric_columns = df.select_dtypes(include=[np.number]).columns

    print("Outlier Detection and Visualization:\n")

    for col in numeric_columns:
        if df[col].isnull().sum() == 0:  
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
            print(f"Feature '{col}' has {len(outliers)} outliers.")

            
            plt.figure(figsize=(10, 5))
            sns.boxplot(data=df, x=col)
            plt.title(f"Boxplot of {col}")
            plt.show()
        else:
            print(f"Feature '{col}' contains missing values and will be skipped.")

else:
    print(f"Error: File not found at '{file_path}'")
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


file_path = "C:dataset.csv"
df = pd.read_csv(file_path)


num_samples = df.shape[0]
num_features = df.select_dtypes(include=['float64', 'int64']).shape[1]

print(f"Number of samples: {num_samples}")
print(f"Number of features: {num_features}")


if num_features > 10:  
    print("The dataset contributes to the curse of dimensionality. Performing PCA...")
    
    
    numeric_features = df.select_dtypes(include=['float64', 'int64'])
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(numeric_features)

    
    pca = PCA(n_components=0.95)  
    principal_components = pca.fit_transform(scaled_data)

    
    print(f"Number of components selected to retain 95% variance: {pca.n_components_}")
    print(f"Explained variance ratio: {pca.explained_variance_ratio_}")
else:
    print("The dataset does not contribute significantly to the curse of dimensionality.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = "C:dataset.csv"
df = pd.read_csv(file_path)


plt.figure(figsize=(8, 6))
plt.scatter(df['bill_length_mm'], df['bill_depth_mm'], alpha=0.7, color='blue')
plt.title('Scatter Plot of Bill Length vs Bill Depth')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.grid(True)
plt.show()


plt.figure(figsize=(8, 6))
plt.plot(df['bill_length_mm'], df['bill_depth_mm'], linestyle='-', color='green', alpha=0.7)
plt.title('Line Plot of Bill Length vs Bill Depth')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.grid(True)
plt.show()


plt.figure(figsize=(8, 6))
plt.hist(df['bill_length_mm'], bins=20, alpha=0.7, color='purple')
plt.title('Histogram of Bill Length')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
plt.hist(df['bill_depth_mm'], bins=20, alpha=0.7, color='orange')
plt.title('Histogram of Bill Depth')
plt.xlabel('Bill Depth (mm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()


sns.boxplot(data=df[['bill_length_mm', 'bill_depth_mm']])
plt.title('Box Plot of Bill Length and Bill Depth')
plt.xlabel('Feature')
plt.ylabel('Value (mm)')
plt.show()


sns.pairplot(df, vars=['bill_length_mm', 'bill_depth_mm'], hue='species', diag_kind='kde', palette='bright')
plt.suptitle('Pairplot of Bill Length and Bill Depth by Species', y=1.02)
plt.show()


correlation_matrix = df[['bill_length_mm', 'bill_depth_mm']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Heatmap of Correlation between Bill Length and Bill Depth')
plt.show()
```

```python
import pandas as pd


file_path = "C:dataset.csv"  
df = pd.read_csv(file_path)  


max_flipper_length = df.groupby(['species', 'island'])['flipper_length_mm'].max().reset_index()



longest_flippers_by_island = max_flipper_length.loc[
    max_flipper_length.groupby('island')['flipper_length_mm'].idxmax()
]

print("Maximum flipper_length_mm for each species and island:")
print(max_flipper_length)


print("\nSpecies with the longest flippers on each island:")
print(longest_flippers_by_island)
```
